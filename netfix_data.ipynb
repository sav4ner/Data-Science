```python
x=dat1.drop(["type"],axis=1)
y=dat1.type
```

```python
x=x.drop(["description"],axis=1)
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder  

le = LabelEncoder()
y=le.fit_transform(y)
y
```

```python
x.country=le.fit_transform(x.country)
x.release_year=le.fit_transform(x.release_year)
x.rating=le.fit_transform(x.rating)
x.duration=le.fit_transform(x.duration)
x.listed_in=le.fit_transform(x.listed_in)
x.title=le.fit_transform(x.title)
x.director=le.fit_transform(x.director)
x.cast=le.fit_transform(x.cast)
x['date_added'] = pd.to_datetime(x['date_added'])
x['date_added'] = pd.to_numeric(x.date_added.replace('-',''))
```

```python
x.date_added
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=109)
```

```python
print(X_train)
```

```python
from sklearn import svm

#Create a svm Classifier= svm.SVC(kernel='linear')
model=svm.SVC()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
print("Accuracy:",accuracy_score(y_test, y_pred))
print("Precision:",precision_score(y_test, y_pred, average='weighted', labels=np.unique(y_pred)))
print("Recall:",recall_score(y_test, y_pred, average='weighted', labels=np.unique(y_pred)))
```
